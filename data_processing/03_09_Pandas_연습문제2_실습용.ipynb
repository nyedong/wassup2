````python
df_covid = pd.read_csv('seoul-covid19_6_30_.csv')
df_covid
````

# Q1-2
```
'연번' 컬럼은 삭제하고  '환자' 컬럼 값 기준으로 정렬한 후 '환자' 컬럼을 인덱스로 설정하여라.
```

````python
df_covid = df_covid.drop('연번', axis=1)
df_covid = df_covid.sort_values(by='환자')
df_covid.set_index('환자')
df_covid
````

# Q1-3
어느 지역에서 제일 많은 확진자가 발생했는지 1~5위 까지 출력하여라 (거주지 기준)

````python
df_top5 = df_covid.groupby('거주지').count().sort_values(by='확진일', ascending = False)
df_top5
````

````python
top5 = list(df_top5['확진일'].nlargest(5, keep='all').index.values)
top5
````


# Q1-4
확진자가 가장 많이 발생한 1, 2, 3위 지역의 데이타만 출력하여라

````python
for i in top5[:3]:
    print(df_covid[df_covid['거주지']==i])
````
