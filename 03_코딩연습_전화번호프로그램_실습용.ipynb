# 전화번호 프로그램 만들기

* 홍길동 010-123-4567
* 기능1 : 연락처 추가
* 기능2 : 연락처 전체 보기
* 기능3 : 검색, 이름을 입력받아서 전화번호 조회
* 기능4 : 수정, 이름을 입력받아서 전화번호 입력수정
* 기능5 : 삭제, 이름 입력받아서 삭제
* 기능6 : 프로그램 종료

````python
tel = '''
1부터 6까지의 숫자를 입력하여 기능을 선택해주세요.
-------------------------------------------------
1. 연락처 추가
2. 연락처 전체 보기
3. 검색
4. 수정
5. 삭제
6. 종료
'''

phone = dict()

while True:  
    user = input(tel)
    
    # 연락처추가
    if user == '1':
        add_name = input('이름 입력: ')
        add_num = input('연락처 입력: ')
        
        # phone.setdefault(add_name, add_num)
        
        if add_name not in phone.keys():
            phone[add_name] = add_num
            print('추가 완료')   
        else:
            print('중복된 이름입니다.')
            
        print('========================================')
        
    # 전체 보기
    elif user == '2':
        for phone_name, phone_num in phone.items():
            print(phone_name, ':', phone_num)
        
        print('=======================================')
        
    # 번호 조회
    elif user == '3':
        name1 = input('번호 조회할 이름 입력: ')
        
        if name1 in phone.keys():
            print(phone[name1])
        else:
            print('입력한 사람의 번호가 없습니다.')
            
        print('=======================================')
        
    # 수정     
    elif user == '4':
        name2 = input('번호 수정할 이름 입력: ')
        
        if name2 not in phone.keys():
            num2 = input('번호 수정: ')
            phone[name2] = num2
            print('수정완료')
        else:
            print('입력한 사람의 번호가 없습니다.')
        
        print('=======================================')
    
    # 삭제
    elif user == '5':
        name3 = input('번호 삭제할 이름 입력: ')
        if name3 in phone.keys():
            del(phone[name3])
            print('삭제완료')
        else:
            print('입력한 사람의 번호가 없습니다.')
        
        print('=======================================')
    
    # 종료
    elif user == '6':
        print('종료')
        break
    
    # 입력 잘못했을 때
    else:
        print('잘못된 입력입니다. 다시 선택해주세요.')
        
        print('=======================================')        
````

contact = {}

while True:
    print('------ 전화번호부 프로그램 ------')
    print('1.추가   2.보기    3.검색   4.수정   5.삭제    9.종료')

    menu = int(input('메뉴를 선택해주세요(숫자 only)'))

    if menu == 1:
        print('연락처를 추가합니다.')
        new_name = input('이름 : ')
        new_tel = input('전화번호 : ')
        print(new_name, new_tel)
        # contact[new_name] = new_tel # 덮어쓰기
        contact.setdefault(new_name, new_tel)

    elif menu == 2:
        print('연락처를 조회합니다.')
        # print(contact)
        # print('='*10)
        for name, tel in contact.items():
            print(name, ":", tel)

    elif menu == 3:
        print('연락처를 검색합니다.')
        search_name = input('검색 이름: ')
        # print(contact[search_name])
        print(contact.get(search_name, '없는 연락처입니다.'))

    elif menu == 4:
        print('연락처를 수정합니다.')
        mod_name = input('수정 이름 : ') # 연락처에 있을 경우에만 새전화번호를 묻는 것이.
        if mod_name in contact:
            mod_tel = input('새 전화번호 : ')
            contact[mod_name] = mod_tel
        else :
            print('등록되지 않은 이름입니다.')

    elif menu == 5:
        print('연락처를 삭제합니다.')
        del_name = input('삭제 이름 : ')
        if del_name in contact:
            del contact[del_name]
        else :
            print('등록되지 않은 이름입니다.')

    elif menu == 9:
        print('프로그램을 종료합니다.') 
        break

    else :
        print('잘못된 입력입니다.')
