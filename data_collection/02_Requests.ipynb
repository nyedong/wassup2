### 상대 URL을 절대 URL로 변환하기

```python
href = sample['href']
```

```python
# //가 중복되는 문제처럼 생각보다 합치는게 번거로울 때가 있음
PATH + href
```

```python
from urllib.parse import urljoin
```

```python
urljoin(PATH, href)
```

```python
# 제목과 url 데이터 목록 만들기
news_title= []
news_url = []

for n in srclist:
    # 태그 객체의 메서드(.txt)를 쓰려면 태그 객체별로 빼줘야함
    title = n.text
    url = urljoin(PATH, n['href'])
    print(title, url)
    news_title.append(title)
    news_url.append(url)
```

```python
news_title
```

```python
news_url
```

### 데이터 프레임 변환

```python
import pandas as pd
```

```python
df = pd.DataFrame({'제목':news_title, '주소':news_url})
df
```

```python
print(df)
```

### excel 파일 저장

```python
# excel 파일 저장
df.to_excel('naver.xlsx', index=False)
```

## 멜론 차트, 노래제목과 가수 수집하기
[멜론 TOP 100 차트 URL](https://www.melon.com/chart/index.htm)

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
PATH = 'http://www.melon.com/chart/index.htm'
resp = requests.get(PATH)
resp
```

```python
html_src = resp.text
html_src
```

```python
# 유저에이전트 정보
info = {'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36'}
```

```python
info
```

```python
requests.get?
```

```python
resp = requests.get(PATH, headers = info)
resp
```

```python
html_src = resp.text
html_src
```

```python
soup = BeautifulSoup(html_src, 'lxml')
soup
```

```python
# 곡이름 찾기
len(soup.select('.ellipsis.rank01 a'))
```

```python
song = [item.text for item in soup.select('.ellipsis.rank01 a')] # item은 태그객체
```

```python
song
```

```python
len(song)
```

```python
# 가수 이름 찾기
# ellpsis rank02로 하면 안됨, 리스트 길이 안맞음
# 가수가 2명이상이거나 동일한 자식속성이 하나 더 있음
artist = []
for item in soup.select('.ellipsis.rank02 a') :
    artist.append(item.text)
print(len(artist))
print(artist)
```

```python
name = [item.text for item in soup.select('.ellipsis.rank02>a')] # > 로 바로 밑 자식만 추가
len(name) # 가수 두명이서 부르면 따로 추가해서 길이 101
```

```python
artist = []
for item in soup.select('.checkEllipsis') :
    artist.append(item.text)
print(len(artist))
print(artist)
```

```python
# 인덱스 만들기
rank_list = list(range(1, 101))
rank_list
```

```python
# 사실 판다스에서 인덱스 핸들링 가능함, 다음에 알아보자
songDF = pd.DataFrame({
        "순위":rank_list,
        "노래제목":song,
        "가수명":artist})
songDF
```

```python
songDF.to_excel('melon100.xlsx', index=False)
```

## rest test site
https://blog.naver.com/gikimirane/222666913565


```python
import requests
```

```python
param = {'name':'Amy', 'age':20, 'address':'Seoul'}
```

```python
# 요청 정보(파라미터)가 인자에 포함되어 있다.
resp1 = requests.get('http://httpbin.org/get', params=param)
print(resp1)
print(resp1.text)
```

```python
# 요청 정보(파라미터)가 인자에 없고, form에 숨겨져있다.(예시라서 보이는 것)
resp2 = requests.post('http://httpbin.org/post', data=param)
print(resp2.text)
```
